```python
## python strings ##

# reminder: python starts counting/indexing at 0
    # values 1,2,3,4 will count like: 0,1,2,3
    # a space itself counts a number as well
# below are examples of slicing 


str = 'Happy 4/20!'
print(str) # prints the complete string 
print(str[0]) # prints first letter of the string
print(str[2:5]) # prints that specified characters, starting at 2 and ending before 5 (2,3,4)
print(str[2:]) # prints whole string not including the start point nor before the start point
print(str[:6]) # prints everthing before the 6, not inlcuding the 6
print(str * 2) # prints string twice
print(str + "TEST") # prints concatenized string
```

```python
## python lists ##

# are variables with multiple varables

list = ['abcd', 786, 2.23, 'john', 70.2]
tinylist = [123, "john"]

print(list)
print(list[0])
print(list[1:3])
print(list[2:])
print(tinylist * 2)
print( list + tinylist) 
```

```python
## python tuples ##

# same as python lists except they can't be modified 
# slicing is the same as before but denoted with () and not []

tuple = ('abcd', 786, 2.23, 'john', 70.2)
tinytuple = (123, 'john')

print(tuple)
print(tuple[0])
print(tuple[1:3])
print(tuple[2:])
print(tinytuple * 2)
print( tuple + tinytuple)
```

```python
## python dictionaries ##

# What it is: step of steps tp perform a task
# Important note: is defined with {}

# How to use it with example below:
dict = {}
dict['one'] = "This is one"
dict[2] = "This is two"
tinydict = {'name' : 'john', 'code' : 6734, 'dept' : 'sales'}

print(dict['one']) # prints value for 'one' key
print(dict[2]) # prints value for 2 key
print(tinydict) # prints complete dictionary
print(tinydict.keys()) # prints all keys for tinydict
print(tinydict.values()) # prints all values for tinydict
```
